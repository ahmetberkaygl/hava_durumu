```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('weatherHistory.csv')
```

```python
df.head(3)
```

```python
df.columns=["bicimlendirilmis_tarih", "ozet", "yagis_turu", "sicaklik", "hissedilen_sicaklik", "nem", "rüzgar_hizi", "ruzgar_yonu", "gorus_mesafesi", "gurultu_ortusu", "basınc", "gunluk_ozet"]
```

```python
df=df[["yagis_turu", "sicaklik", "hissedilen_sicaklik", "nem", "rüzgar_hizi", "ruzgar_yonu", "gorus_mesafesi", "gurultu_ortusu", "basınc"]]
```

```python
df.head(3)
```

```python
df.yagis_turu.unique()
```

```python
df = df.dropna()
```

```python
df.yagis_turu.unique()
```

```python
df["yagis_turu"] = df.yagis_turu.map({"rain":1,"snow":2})
```

```python
df.head(10)
```

```python
y=df[["yagis_turu"]]
x=df.drop("yagis_turu",axis=1)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.8,random_state=42)
```

```python
lm = LinearRegression()
model = lm.fit(x_train,y_train)
model.score(x_test,y_test)
```

```python
model.predict([[-1.11,-1.150000,	0.90,	0,	0,	0.4830,	0.0,	1018.56]])
```
